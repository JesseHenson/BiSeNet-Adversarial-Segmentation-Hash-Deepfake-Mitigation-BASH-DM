```python
import glob
import cv2
import matplotlib.pyplot as plt
import pandas as pd

segmented_dir = '/Users/jesse/Documents/Segmentation_Adversarial_Watermarking_for_Deepfakes/results/cropped_segments/0001'
file_arr = []
percent_useful_arr = []

for filename in glob.glob(f'{segmented_dir}/*.png'):
    img = cv2.imread(filename,cv2.IMREAD_GRAYSCALE)
    im_bw = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]
    percent_useful = sum(sum(im_bw/255))/(32*32)
    # print(f'file: {filename[-6:-4]} - usable preportion: {percent_useful}')
    file_arr.append(filename[-6:-4])
    percent_useful_arr.append(percent_useful)

df = pd.DataFrame({'file_name':file_arr,'percent_usage':percent_useful_arr})
display(df.sort_values('percent_usage'))
```

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('results/cropped_segments/0001/ref/Adam_Sandler_0001_01.png')
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
im_bw = cv2.threshold(img_grey, 127, 255, cv2.THRESH_BINARY)[1]
# plt.imshow(cv2.cvtColor(img[0:32,0:32], cv2.COLOR_BGR2RGB))
max_useful = 0
max_img = []
for section_x in range(int(512/16)):
    for section_y in range(int(512/16)):
        print(section_x, section_y)
        percent_useful = sum(sum(im_bw[section_x*16:32+(section_x*16),section_y*16:32+(section_y*16)]/255))/(32*32)
        # print(percent_useful)
        # percent_useful_arr.append(percent_useful)
        if percent_useful > max_useful:
            max_useful = percent_useful
            max_img = img[section_x*16:32+(section_x*16),section_y*16:32+(section_y*16)]
    
print(max_useful)
plt.imshow(cv2.cvtColor(max_img, cv2.COLOR_BGR2RGB))
```

```python
import cv2
import numpy as np

# mask = cv2.imread('mask.png', 0)
bg = cv2.imread('lfw-deepfunneled/Adam_Sandler/Adam_Sandler_0001.jpg')
fg = cv2.imread('results/pictures/file1_i1_predicted1_encoded1_truth71.png')

mask = 255*np.ones(fg.shape[:2], fg.dtype)

top, bottom, left, right = 20, 20, 20, 20
border_bg = cv2.copyMakeBorder(bg, top, bottom, left, right, cv2.BORDER_CONSTANT, value=[0, 0, 0])
border_fg = cv2.copyMakeBorder(fg, top, bottom, left, right, cv2.BORDER_CONSTANT, value=[0, 0, 0])
border_mask = cv2.copyMakeBorder(mask, top, bottom, left, right, cv2.BORDER_CONSTANT, value=0)
grey_img = cv2.cvtColor(fg, cv2.COLOR_BGR2GRAY)
(thresh, binary_img) = cv2.threshold(grey_img, 127, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

c = max(contours, key=cv2.contourArea)
x, y, w, h = cv2.boundingRect(c)

center = (int(x + w * 0.5), int(y + h * 0.5))

output = cv2.seamlessClone(border_bg, border_fg, border_mask, p=center, flags=cv2.NORMAL_CLONE)

# cv2.imshow("border_mask", border_mask)
# cv2.imshow("border_bg", border_bg)
# cv2.imshow("border_fg", border_fg)
# cv2.imshow('output', output)

# plt.imshow(cv2.cvtColor(border_mask, cv2.COLOR_BGR2RGB))
plt.imshow(cv2.cvtColor(border_bg, cv2.COLOR_BGR2RGB))
plt.imshow(cv2.cvtColor(border_fg, cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(output,cv2.COLOR_BGR2RGB))
```
